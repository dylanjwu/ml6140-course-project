```python
import matplotlib.pyplot as plt

X = ['BoW', 'BoW-chars', 'TF-IDF', 'TF-IDF-CHARS']
y1 = [0.58, 0.43, 0.61, 0.44]  # Naive Bayes mean scores
y2 = [0.60, 0.52, 0.58, 0.50]  # Decision Tree mean scores

plt.plot(X, y1, color='red', label='NB mean score')  # Naive Bayes line
plt.plot(X, y2, color='green', label='DT mean score')  # Decision Tree line

plt.xlabel('Vectorization Methods')
plt.ylabel('Mean CV Scores')
plt.title('Comparison of Mean Scores: Naive Bayes vs. Decision Tree')
plt.legend()
plt.grid(True)

plt.show()
```
