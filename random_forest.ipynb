```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split


from models import run_model, get_data
```

```python
filename = '/Users/dylanwu/Development/cs_6140/project/data/tfidf_pca_100features_1000rows.csv'
X, y = get_data(filename)

oob_errors = []
accuracies = []

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)  # Adjust test_size as needed


for num_trees in range(50, 200):
    print(f'i: {num_trees - 20}')
    rf = RandomForestClassifier(n_estimators=num_trees, oob_score=True, random_state=42)

    rf.fit(X_train, y_train)

    # Access the out-of-bag error (OOB score)
    oob_error = 1 - rf.oob_score_
    print(f'oob_error: {oob_error:.4f}')
    oob_errors.append(oob_error)

    # Calculate accuracy for the current model
    y_pred = rf.predict(X_test)  # Predictions on the training data
    accuracy = accuracy_score(y_test, y_pred)
    print(f'accuracy: {accuracy:.4f}')
    accuracies.append(accuracy)

print(oob_errors)
print(accuracies)
```

```python
plt.plot(range(50, 200), oob_errors, marker='o', linestyle='-', color='blue', linewidth=1)
plt.plot(range(50, 200), accuracies, marker='o', linestyle='-', color='green', linewidth=1)
plt.xlabel('Number of trees')
plt.ylabel('Accuracy (green), OOB Error (blue)')
plt.grid(False)  # Add gridlines

# Show the plot
plt.show()
```
